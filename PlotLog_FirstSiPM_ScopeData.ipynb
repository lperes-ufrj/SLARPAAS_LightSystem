```python
#Bias connected directed in the SiPM PCB
csv_file = 'FirstSiPM_ScopeData/scope1Mohm-pamp3.csv'

# Read the CSV file into a pandas DataFrame
data = pd.read_csv(csv_file,skiprows=1)
data.head()

# Extract the columns
x = data.iloc[:, 0]*10**6  # the first column is seconds
y1 = data.iloc[:, 1]  # the second column is Volts (WaveGenerator)
y2 = data.iloc[:, 2]  # the second column is Volts (SiPM)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x, y1, marker='.', linestyle='--', label='WaveGenerator')
plt.plot(x, y2, marker='.', linestyle='--',  label='SiPM')
plt.ylabel('Volts (V)')
plt.xlabel(r'micro seconds ($\mu$s)')
#plt.yscale('log')
plt.title('Plot of CSV Data Oscilloscope')
plt.legend(loc='upper left', title = f'Max. Amplitude Signal SiPM: {y2.max()-y2.min()} V\nMax. Vpp Amplitude WaveGenerator: {(y1.max()-y1.min())/2} V', ncols=2 )
plt.show()
```

```python
import scipy.optimize as opt
import numpy as np

def model_expodecay(t, A, K, C):
    return A * np.exp(-K * t) + C

opt_parms, parm_cov = opt.curve_fit(model_expodecay, x[x>100], y2[x>100], p0=[1.,0.01, 1.2])
A, K, C = opt_parms
```

```python
print(1./K) ## ~15 mu s characteristic of which process, what is happing here???
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x[x>100], y2[x>100], marker='.', linestyle='--',  label='SiPM')
plt.plot(x[x>100],model_expodecay(x[x>100],A,K,C))
plt.plot()
plt.ylabel('Volts (V)')
plt.xlabel(r'micro seconds ($\mu$s)')
#plt.yscale('log')
plt.title('Plot of CSV Data Oscilloscope')
plt.legend(loc='upper left', title = f'Max. Amplitude Signal SiPM: {y2.max()-y2.min()} V\nMax. Vpp Amplitude WaveGenerator: {(y1.max()-y1.min())/2} V', ncols=2 )
plt.show()

### Ok, the fit is good enough
```
